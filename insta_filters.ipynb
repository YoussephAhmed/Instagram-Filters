## Imports 

```python
import cv2
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import numpy as np
```

## Read the input Image 

```python
img = cv2.imread('Joe.jpg',1)

blue  = img[:, :, 0]
green = img[:, :, 1]
red   = img[:, :, 2]

img = cv2.merge((red,green,blue))

plt.imshow(img)


```

# 1) Inkwell Filter
### black and white (gray) filter

```python
def Inkwell_filter(img):

    new_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    # alpha for contrast
    # beta for brightness
    new_img = cv2.convertScaleAbs(new_img, alpha=1 , beta = -20)
    plt.imshow(new_img,cmap='gray')
```

```python
Inkwell_filter(img)
```

# 2) X-Pro 11 Filter
### bumps up the contrast

```python
def X_Pro_11_filter(img):

    # alpha for contrast
    # beta for brightness
    new_img = cv2.convertScaleAbs(img, alpha=1.3 , beta = -50)
    plt.imshow(new_img)
```

```python
X_Pro_11_filter(img)
```

# 3) Lily Filter
### greenish filter

```python
def Lily_filter(img):

    red   = img[:, :, 0]
    green = img[:, :, 1]
    blue  = img[:, :, 2]
    
    
    new_blue  = ((blue / 255) ** 1 ) * 255
    new_green = ((green / 255) ** 0.5 ) * 255
    new_red  = ((red / 255) ** 1 ) * 255

    new_blue = np.uint8(new_blue)
    new_green = np.uint8(new_green)
    new_red = np.uint8(new_red)


    new_img = cv2.merge((new_red,new_green,new_blue))

    plt.imshow(new_img)
    
```

```python
Lily_filter(img)
```

```python
_1977_filter(img)
```
